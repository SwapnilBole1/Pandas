### Addition using append

```python
import pandas as pd
df = pd.DataFrame([[1, 2], [3, 4]], columns = ['a','b'])
df2 = pd.DataFrame([[5, 6], [7, 8]], columns = ['a','b'])
```

```python
df #dataset1
```

```python
df2 #dataset2
```

```python
ndf = df.append(df2,ignore_index=True)
```

```python
ndf
```

```python
pd.concat([df,df2],axis =1,ignore_index=True)
```

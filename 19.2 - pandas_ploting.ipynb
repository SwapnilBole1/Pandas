# Plotting using Pandas

```python
import numpy as np
import pandas as pd
```

```python
data = np.random.rand(4,4)
```

```python
data
```

```python
df = pd.DataFrame(data,columns=list('abcd'))
df.head(2)
```

# Line plot

```python
#all columns to be considerd
df.plot(kind='line')
```

```python
df.plot() #default its line plot
```

```python
# only 2 columns plot we want
df[['b','d']].plot(kind='line')
```

```python
import matplotlib.pyplot as plt
df[['b','d']].plot()
plt.title('b and d plot')
```

```python
help(df.plot)
```

```python
df.plot()
```

```python
df.plot(subplots = True, figsize=(12,8),sharex=False)
```

```python
# Now i dont want combined plot, we need subplots
ax1 = df.plot(figsize=(12,8),sharex=False)
ax1.legend(loc=2,fontsize=20)
ax1.set_ylabel('kw',fontdict={'fontsize':24})
plt.xlabel('X values', fontname="DejaVu Sans", fontsize=20)
plt.tight_layout()
```

```python
#change layout
df.plot(subplots=True,layout=(2,2),title='Combined plot',figsize=(10,7))
plt.show()
```

# Scatter plot

```python
df.plot(x = 'a',y ='c',kind='scatter')
```
